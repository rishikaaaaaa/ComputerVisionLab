```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import matplotlib.image as mpimg
```

```python
# Define the size of the image in rows and columns
rows = 100
cols = 100

# Create a 2D NumPy array with a diagonal line separating white and black pixels
image = np.zeros((rows, cols, 3), dtype=np.uint8)

# Set the upper part of the diagonal to white (255, 255, 255) and the lower part to black (0, 0, 0)
for i in range(rows):
    for j in range(cols):
        if i < j:
            image[i, j] = [255, 255, 255]  # White pixel
        else:
            image[i, j] = [0, 0, 0]  # Black pixel

# Print the array
print(image)

# Display the image using Matplotlib
plt.imshow(image)
plt.axis('on')
plt.show()
```

```python
# Define the size of the chessboard in rows and columns
rows = 8
cols = 8

# Create an array of 1 and 0 alternately
array = np.zeros((rows, cols), dtype=int)

for i in range(rows):
    for j in range(cols):
        if (i + j) % 2 == 1:
            array[i, j] = 1

# Display the array
print("Array:")
print(array)

# Display the chessboard pattern
plt.imshow(array, cmap='gray', interpolation='nearest', extent=(0, cols + 0, rows + 0, 0))
plt.axis('on')
plt.show()
```

```python
import cv2
# Read the image using OpenCV
image_3 = 'D:/3.png'
image_three = cv2.imread(image_3)
image_three_r = cv2.resize(image_three, (1000,750))

image_2 = 'D:/2.jpg'
image_two = cv2.imread(image_2)
image_two_r = cv2.resize(image_two, (1000,750))

plt.figure(figsize = (10,10))

plt.subplot(1,2,1)
plt.imshow(image_three_r)
plt.axis('on')
plt.title('three')

plt.subplot(1,2,2)
plt.imshow(image_two_r)
plt.axis('on')
plt.title('two')
```

```python
def translate_image(image, dx, dy):

    rows, cols, channels = image.shape

    # Create an empty image for the translated result
    translated_image = np.zeros_like(image, dtype=np.uint8)

    # Apply translation manually
    for i in range(rows):
        for j in range(cols):
            new_i = i + dy
            new_j = j + dx

            # Check if the new indices are within bounds
            if 0 <= new_i < rows and 0 <= new_j < cols:
                translated_image[new_i, new_j] = image[i, j]

    return translated_image

# Define translation parameters (dx and dy)
dx = -350
dy = 0

# Translate the image
translated_image_1 = translate_image(image_three_r, dx, dy)

# Display the original and translated images side by side
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_three_r, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(translated_image_1, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')
plt.axis('off')

plt.show()
```
