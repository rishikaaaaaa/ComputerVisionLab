```python
# import OpenCV and pyplot 
import cv2 
from matplotlib import pyplot as plt 
import numpy as np
```

```python
# read left and right images 
imgR = cv2.imread("D:/images/im0.png", cv2.IMREAD_GRAYSCALE) 
imgL = cv2.imread("D:/images/im1.png", cv2.IMREAD_GRAYSCALE) 

# creates StereoBm object 
stereo = cv2.StereoBM_create(numDisparities = 64, blockSize = 15) 

# computes disparity 
disparity = stereo.compute(imgR, imgL) 

# Convert disparity to depth map
baseline = 1 # Example: baseline distance between cameras in meters
focal_length = 1  # Example: focal length of the cameras in meters

depth_map = baseline * focal_length / disparity

plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
plt.imshow(imgR, cmap='gray')
plt.title('Right')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imgL, cmap='gray')
plt.title('Left')
plt.axis('off')
plt.show()

plt.figure(figsize=(15,5))
# disparity map 
plt.subplot(1, 2, 1)
plt.imshow(disparity, 'gray') 
plt.title('Disparity Map')
plt.axis("off")


# Plot the depth map
plt.subplot(1, 2, 2)
plt.imshow(depth_map, cmap = 'gray')
plt.title('Depth Map')
plt.axis("off")
plt.show()
```
