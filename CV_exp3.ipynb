```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read input image
img = cv2.imread('D:/erosion.png', 0)

# Define structuring element
kernel = np.ones((5,5), np.uint8)

# Perform erosion
erosion = cv2.erode(img, kernel, iterations = 2)

# Perform dilation
dilation = cv2.dilate(img, kernel, iterations = 2)

plt.figure(figsize=(15, 8))
plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(erosion, cmap='gray')
plt.title('Image after Erosion')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(dilation, cmap='gray')
plt.title('Image after Dilation')
plt.axis('off')

plt.show()
```

```python
img2 = cv2.imread('D:/thumb.png', 0)

def erosion(img2, SE):
  imgErode= cv2.erode(img2,SE,2)
  return imgErode

#Function for dilation
def dilation(img2, SE):
  imgDilate= cv2.dilate(img2,SE,1)
  return imgDilate

SE= cv2.getStructuringElement(cv2.MORPH_RECT,(3,3)) #Define the structuring element using inbuilt CV2 function
AeB= erosion(img2,SE) #Erode the image
AoB= dilation(AeB, SE) #Dilate the eroded image. This gives opening oepration
AoBdB= dilation(AoB,SE) #dilate the opened image followed by ersoion. This will give closing of the openeed image
AoBdBeB= erosion(AoBdB, SE)

plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.imshow(img2, cmap="gray")
plt.title("Original")
plt.axis('off')

plt.subplot(3,2,2)
plt.title("Erosion")
plt.imshow(AeB, cmap="gray")
plt.axis('off')

plt.subplot(3,2,3)
plt.title("Opening Operation")
plt.imshow(AoB, cmap="gray")
plt.axis('off')

plt.subplot(3,2,4)
plt.title("Further Dilation")
plt.imshow(AoBdB, cmap="gray")
plt.axis('off')

plt.subplot(3,2,5)
plt.title("Closing Operation")
plt.imshow(AoBdBeB, cmap="gray")
plt.axis('off')
```

```python
# Load the image (replace 'your_image.jpg' with your image file)
image = cv2.imread('D:/ct scan.png', cv2.IMREAD_GRAYSCALE)

# Define a 3x3 structuring element
kernel = np.ones((3, 3), np.uint8)

# Grayscale Dilation
dilation = cv2.dilate(image, kernel, iterations=1)

# Grayscale Erosion
erosion = cv2.erode(image, kernel, iterations=1)

# Morphological Gradient
gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)

# Display the original image and processed images
plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(dilation, cmap='gray')
plt.title('Dilation')
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(erosion, cmap='gray')
plt.title('Erosion')
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(gradient, cmap='gray')
plt.title('Morphological Gradient')
plt.axis("off")

plt.show()
```
